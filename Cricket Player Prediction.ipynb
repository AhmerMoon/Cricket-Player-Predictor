```python
#training and testing data
from sklearn.model_selection import train_test_split

X = demo.drop(['Runs'], axis=1)
Y = demo['Runs']
comp = dict()

train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.20, shuffle=True)

```

```python
from sklearn.tree import DecisionTreeRegressor
import difflib

def find_best_match(user_input, valid_options):
    """Find best match ignoring case, spaces etc. Returns None if no good match found."""
    user_input_clean = user_input.lower().replace(" ", "")
    # Clean valid options similarly
    cleaned_options = {opt: opt.lower().replace(" ", "") for opt in valid_options}
    # Find close matches based on cleaned names
    matches = difflib.get_close_matches(user_input_clean, cleaned_options.values(), n=1, cutoff=0.6)
    if matches:
        # Return original column name corresponding to the cleaned match
        for orig, clean in cleaned_options.items():
            if clean == matches[0]:
                return orig
    return None

# Get all player/opposition columns (excluding known numeric columns)
numeric_cols = {'Runs', 'BF', 'Overs'}
valid_cols = [col for col in final.columns if col not in numeric_cols]

player_input = input('Enter the name (e.g., Rohit Sharma): ').strip()
opposition_input = input('Enter opposition team (e.g., England): ').strip()

# Find best match column names
player_col = find_best_match(player_input, valid_cols)
opposition_col = find_best_match(opposition_input, valid_cols)

if player_col is None:
    raise ValueError(f"Player name '{player_input}' not found in data columns.")
if opposition_col is None:
    raise ValueError(f"Opposition team '{opposition_input}' not found in data columns.")

bf = float(input('Enter balls faced: '))
ov = float(input('Enter the overs: '))

# Extract relevant data subset
demo = final[[player_col, opposition_col, 'Runs', 'BF', 'Overs']]
X = demo.drop(['Runs'], axis=1)
Y = demo['Runs']

# Train the model
tree = DecisionTreeRegressor()
tree.fit(X, Y)

# Prepare input vector for prediction (1 for player and opposition presence)
preds = tree.predict([[1, 1, bf, ov]])
print(f"{player_col}'s overall run predicted is {int(preds[0])} against {opposition_col}.")
```
